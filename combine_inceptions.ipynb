# **Abnormal**

```
# load train, test,  valid data 
label_type = "abnormal"
exam_view ="axial"
data_path = "/content/gdrive/My Drive/Colab Notebooks/MRNet/dataset/MRNet-v1.0"
p, l = partition(data_path)

with open(f'/content/gdrive/My Drive/Colab Notebooks/MRNet/intermediate_dataset/train-abnormal.csv', newline='') as csvfile:
      data_train = list(csv.reader(csvfile))
x_train = np.array(data_train,dtype='float32')
x_train  = x_train[:,1:4]

with open(f'/content/gdrive/My Drive/Colab Notebooks/MRNet/intermediate_dataset/test-abnormal.csv', newline='') as csvfile:
      data_test = list(csv.reader(csvfile))
x_test = np.array(data_test,dtype='float32')
x_test  = x_test[:,1:4]

with open(f'/content/gdrive/My Drive/Colab Notebooks/MRNet/intermediate_dataset/valid-abnormal.csv', newline='') as csvfile:
      data_valid = list(csv.reader(csvfile))
x_valid = np.array(data_valid,dtype='float32')
x_valid  = x_valid[:,1:4]

# load labels
ids_train = sorted(p['train'][exam_view]);
ids_test = sorted(p['test'][exam_view]);
ids_valid = sorted(p['valid'][exam_view]);

arr = [] 
for e in ids_train:
  arr.append(l[e][label_type]);

y_train = np.array(arr,dtype='float32').reshape(len(arr), 1);

arr = [] 
for e in ids_test:
  arr.append(l[e][label_type]);

y_test = np.array(arr,dtype='float32').reshape(len(arr), 1);

arr = [] 
for e in ids_valid:
  arr.append(l[e][label_type]);
  
y_valid = np.array(arr,dtype='float32').reshape(len(arr), 1);

# class weight for skewed classes
pos = 0
for e in p['train'][exam_view]:
  if l[e][label_type] == 1:
    pos += 1
total = len(p['train'][exam_view])
neg = total - pos
class_weight = {0: 1, 1: neg/pos}
if neg < pos:
  class_weight = {0: pos/neg, 1: 1}


model,cbs = LR_model(0.001,label = "abnormal",modelName="inception");
model.fit(x=x_train,y=y_train, validation_data =[x_test,y_test] ,epochs=100, initial_epoch = 0,verbose=1, class_weight=class_weight,callbacks=[cbs[0]],shuffle=False, use_multiprocessing=True, workers=6)
```

```

model,cbs = LR_model(0,label = "abnormal",modelName="inception");
model.load_weights('/content/gdrive/My Drive/Colab Notebooks/MRNet/training_inception_LR/abnormal/weights.100.hdf5');
metrics = model.evaluate(x=x_valid,y=y_valid,verbose=1);
print(metrics)
```
